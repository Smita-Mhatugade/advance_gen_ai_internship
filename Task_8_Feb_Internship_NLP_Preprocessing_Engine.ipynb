## **Task 1: Conceptual Understanding:**

**1. What is the difference between "Love" and "love" in NLP?**

In Natural Language Processing (NLP), the difference between "Love" and "love" is mainly due to case sensitivity. "Love" (with a capital L) may be treated as a different token and can represent the beginning of a sentence or even a proper noun (such as a name). On the other hand, "love" (in lowercase) is generally interpreted as a regular word expressing the emotion of affection.

However, in many NLP applications, a preprocessing step called lowercasing is applied, where all words are converted to lowercase. In such cases, both "Love" and "love" are treated as the same word, and the difference between them is ignore

**2. What happens if stopwords are not removed?**

If stopwords are not removed in NLP, the text will contain many common words like “the”, “is”, “and”, which usually do not add much meaning.

can lead to:

More noise in the data → Important words get less focus

Increased computation → More words to process, making models slower

Lower model performance (sometimes) → Especially in tasks like text classification or search

**3 .Provide two real-world scenarios where removing stopwords can be harmful.**

Sentiment Analysis:

Stopwords like “not”, “no”, “never” are crucial for understanding sentiment. For example, the sentence “I do not like this product” would become “I like this product” after removing stopwords, completely changing the meaning.
Machine Translation

Language Modeling:

Stopwords help maintain the grammatical structure of a sentence. Removing them can make sentences incomplete or confusing, leading to incorrect translations or poor language generation.

**4. Explain the difference between stemming and lemmatization.**

Stemming is a simple method that just cuts off word endings. It doesn’t care about the actual meaning, so sometimes the result may not be a real word.

Example: “running” → “runn”

Lemmatization is smarter. It looks at the meaning of the word and converts it into a proper base word found in a dictionary.

Example: “running” → “run”, “better” → “good”

## **Task 2:Build Advanced Preprocessing Function**

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
import re
```

```python
import re

def preprocess_text(text):
  # 1. Remove numbers
  text = re.sub(r'\d+', '', text)

  # 2. Remove extra spaces (initial normalization)
  text = re.sub(r'\s+', ' ', text).strip()

  # 4. Convert text to lowercase
  text = text.lower()

  # Remove punctuation and other non-alphabetic characters (implied by examples)
  text = re.sub(r'[^a-z\s]', '', text)

  # Re-normalize spaces after punctuation removal, before splitting
  text = re.sub(r'\s+', ' ', text).strip()

  words = text.split()
  processed_words = []
  exceptions_for_short_tokens = ['no', 'not']

  for word in words:
      # 3. Handle repeated characters
      word = re.sub(r'(.)\1{2,}', r'\1', word)

      # 5. Remove very short tokens
      if len(word) > 2 or word in exceptions_for_short_tokens:
          processed_words.append(word)

  text = ' '.join(processed_words)

  # 6. Remove URLs and email-like patterns
  text = re.sub(r'https?://\S+|www\.\S+|\S+@\S+', '', text)

  # Final space normalization
  text = re.sub(r'\s+', ' ', text).strip()

  return text
```

## **Task 4: Token Analytics**

```python
analytics_results = []

for i, text in enumerate(sample_inputs):
    cleaned_sentence = preprocess_text(text)
    cleaned_tokens = cleaned_sentence.split()

    total_tokens = len(cleaned_tokens)
    unique_tokens = len(set(cleaned_tokens))

    average_token_length = 0
    if total_tokens > 0:
        average_token_length = sum(len(token) for token in cleaned_tokens) / total_tokens

    analytics_results.append({
        "original_text": text,
        "cleaned_sentence": cleaned_sentence,
        "cleaned_tokens": cleaned_tokens,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "average_token_length": f"{average_token_length:.2f}"
    })

print("--- Token Analytics ---")
for i, result in enumerate(analytics_results):
    print(f"\n--- Sentence {i+1} ---")
    print(f"Original: {result['original_text']}")
    print(f"Cleaned: {result['cleaned_sentence']}")
    print(f"Total Tokens: {result['total_tokens']}")
    print(f"Unique Tokens: {result['unique_tokens']}")
    print(f"Average Token Length: {result['average_token_length']}")


print("\n--- Analysis Questions ---")
print("\nWhich sentence had the most noise?")
print("Based on the reduction in complexity and removal of non-alphabetic characters, numbers, and excessive repetitions, sentences like 'Get 100% FREE access now!!!' (due to numbers, punctuation) and 'I absolutely looooved this product 😍😍' (due to emojis, repeated characters) had a high degree of 'noise' that was effectively cleaned. 'Nooooo this is baaad!!!' also had significant noise from repetition and punctuation. The sentence 'Call me at 9876543210' was reduced to just 'call', indicating almost all its original content was noise in an NLP context focused on words.")

print("\nWhich sentence retained the most meaningful tokens after cleaning?")
print("Sentences that were already relatively clean or where the noise didn't heavily obscure the core message tended to retain more meaningful tokens. For example, 'I am not happy with this' (which became 'not happy with this') retained its core sentiment-bearing words. 'This is THE best course!!!' (cleaned to 'this the best course') also kept its essential meaning intact. The sentences where stopwords like 'I' and 'is' were removed but 'no' and 'not' were preserved, like 'Nooooo this is baaad!!!' ('no this bad'), also effectively retained their primary meaning due to the exceptions for short tokens.")
```

## **Task 5: Frequency Analysis**

```python
from collections import Counter

# Combine all cleaned tokens from the analytics_results
all_cleaned_tokens = []
for result in analytics_results:
    all_cleaned_tokens.extend(result['cleaned_tokens'])

# Perform frequency analysis
word_counts = Counter(all_cleaned_tokens)

print("--- Frequency Analysis ---")

# Top 10 most frequent words
print("\nTop 10 Most Frequent Words:")
for word, count in word_counts.most_common(10):
    print(f"'{word}': {count}")

# Top 5 least frequent words

print("\nTop 5 Least Frequent Words:")
# Sort by count ascending, then by word alphabetically for consistent output
least_common = sorted(word_counts.items(), key=lambda item: (item[1], item[0]))
for i, (word, count) in enumerate(least_common):
    if i < 5: # Get the first 5 after sorting
        print(f"'{word}': {count}")
```

## **Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        # Preprocess each text using the previously defined function
        cleaned_sentence = preprocess_text(text)

        # Split the cleaned sentence into tokens
        tokens = cleaned_sentence.split()

        # Extend the list of all tokens
        all_tokens.extend(tokens)

        # Append the cleaned sentence
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
