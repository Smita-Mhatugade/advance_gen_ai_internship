```python
#Problem 1: Smart Light Controller
#Problem Statement
#Create a class to control a smart light that can be turned ON or OFF and display its status.
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")
light = SmartLight("Bedroom Light")
action = input("Enter Action (ON/OFF): ")
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
#Problem 2: Employee ID Card System
#Problem Statement:Create a class that stores employee name, ID, and department and displays ID card details.
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
emp = Employee(name, emp_id, department)
emp.display_id_card()
```

```python
#Problem 3: Mobile Contact Record
#Problem Statement: Create a class to store contact name and phone number and display contact information.
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
name = input("Contact Name: ")
phone = input("Phone Number: ")
c1 = Contact(name, phone)
c1.display_contact()
```

```python
#Problem 4: Product Price Tag Generator
#Problem Statement: Create a class that stores product name and price and prints a formatted price tag.
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))

name = input("Product Name: ")
price = input("Price: ")
p1 = Product(name, price)
p1.display_price_tag()
```

```python
#Problem 5: Movie Rating Display System
#Problem Statement:Create a class that stores movie name and rating and displays movie details.
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("Movie Name:", self.name)
        print("Rating:", self.rating)

name = input("Movie Name: ")
rating = input("Rating: ")
m1 = Movie(name, rating)
m1.display_details()
```
