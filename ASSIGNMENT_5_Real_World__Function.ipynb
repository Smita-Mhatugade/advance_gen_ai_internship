Problem 1: Social Media – Post Engagement Analyzer
Problem Statement
Create a function to analyze engagement on a social media post.
Rules
Calculate total likes
If Total Likes ≥ 1000 → Viral Post
Otherwise → Normal Engagement


```python
def post_engagement_analyzer():
    likes_list = []
    n = int(input("Enter number of likes entries: "))

    for i in range(n):
        likes = int(input(f"Enter likes {i+1}: "))
        likes_list.append(likes)

    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)
post_engagement_analyzer()
```

Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement
Create a function that analyzes seasonal rainfall data.
Rules
Calculate average rainfall
Average ≥ required level → Adequate Rainfall
Otherwise → Inadequate Rainfall


```python
def rainfall_checker():
    rainfall_list = []
    n = int(input("Enter number of days: "))

    for i in range(n):
        rainfall = float(input(f"Enter rainfall for day {i+1}: "))
        rainfall_list.append(rainfall)

    required = float(input("Enter required rainfall level: "))

    total = 0
    for r in rainfall_list:
        total += r

    average = total / n

    if average >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall_checker()
```

Problem 4: Social Media – Duplicate Account Detection
Problem Statement
Create a function to detect duplicate usernames in a system.
Rules
If duplicates exist → Flag accounts
Else → Accounts are unique
Real-Time Use
Platform moderation
Fraud detection systems
Hint
Use function
Use set data structure
Expected Output Format
Duplicate Accounts Found: Yes


```python
def duplicate_account_detection():
    usernames = []
    n = int(input("Enter number of usernames: "))

    for i in range(n):
        user = input(f"Enter username {i+1}: ")
        usernames.append(user)

    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

duplicate_account_detection()
```

Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement
Create a function that filters crop prices above a premium threshold.
Rules
Price > ₹2000 → Premium Crop
Else → Regular Crop
Real-Time Use
Market analytics
Export pricing systems
Hint
Use function
Use list filtering logic
Expected Output Format
Premium Crops: [2500, 3200]


```python
def premium_crop_filter():
    prices = []
    n = int(input("Enter number of crop prices: "))

    for i in range(n):
        price = int(input(f"Enter price {i+1}: "))
        prices.append(price)

    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)

premium_crop_filter()
```

Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
Rules
Attendance ≥ 75% → Eligible
Attendance < 75% → Not Eligible
Real-Time Use
Learning management systems
Internship tracking platforms
Hint
Use function
Use list and loop
Expected Output Format
Attendance Percentage: 80.0
Exam Eligibility: Eligible

```python
def attendance_checker():
    attendance = []
    n = int(input("Enter number of days: "))

    for i in range(n):
        day = int(input(f"Enter attendance for day {i+1} (1=Present, 0=Absent): "))
        attendance.append(day)

    total_present = 0
    for a in attendance:
        total_present += a

    percentage = (total_present / n) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_checker()
```

Problem 10: Smart Electricity Bill Analyzer
Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing Rules
First 100 units → ₹3 per unit
Next 100 units (101–200) → ₹5 per unit
Above 200 units → ₹7 per unit
Usage Classification
Total bill < ₹500 → Low Usage
₹500 – ₹1500 → Moderate Usage
₹1500 → High Usage
Real-Time Use
Smart meter systems
Electricity board billing software
Energy consumption analytics
Hint
Use function to calculate bill
Use conditional logic for slabs
Use return values (bill and usage status)


```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


def main():
    units = int(input("Enter units consumed: "))
    bill, status = electricity_bill(units)

    print("Total Bill:", bill)
    print("Usage Status:", status)
main()
```
