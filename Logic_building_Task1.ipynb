# User Login Check

Problem Statement:
Given a username and password, check whether login is successful.

username = "admin"

password = "1234"




```python
username = "admin"
password = "1234"

username_input = input("Enter your username = ")
password_input = input("Enter your password = ")

if username_input == username and password_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# **Pass / Fail Analyzer**
Problem Statement:
Analyze student results.

marks = [45, 78, 90, 33, 60]
Requirements:

A student passes if marks ≥ 50

Count the total number of pass students

Count the total number of fail students

Print the final result clearly



```python
marks = [45, 78, 90, 33, 60]

def pass_fail(marks):
    pass_students = 0
    fail_students = 0

    for mark in marks:
        if mark >= 50:
            pass_students += 1
        else:
            fail_students += 1

    print("Total Students =", len(marks))
    print("Pass Students =", pass_students)
    print("Fail Students =", fail_students)

pass_fail(marks)
```

# **Simple Data Cleaner**
Problem Statement:
Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]

Requirements:

Remove extra spaces

Convert all names to lowercase

Print the cleaned list


```python
names = [" Alice ", "bob", " CHARLIE "]

def clean_data(names):
    cleaned = [name.strip().lower() for name in names]
    print(cleaned)

clean_data(names)
```

# **Message Length Analyzer**

Problem Statement:

Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:

Print the length of each message

Flag messages longer than 10 characters


```python
messages = ["Hi", "Welcome to the platform", "OK"]

def message_length(messages):
    for msg in messages:
        length = len(msg)
        print(f"Message: '{msg}' → Length: {length}")

        if length > 10:
            print("⚠ This message is longer than 10 characters")

message_length(messages)
```

# **Error Message Detector**
Problem Statement:

Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:

  Count the number of "ERROR" entries

  Print the total error count


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

def error_message(logs):
    error_count = logs.count("ERROR")
    print(f"The total error count is {error_count}")

error_message(logs)
```
